```python
for feature in numerical_features:
    data1=data.copy()
    data1.groupby(feature)['turnover'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('turnover')
    plt.title(feature)   # here we see the relationship between value and all other parameters
    plt.show()
```

```python
for feature in numerical_features:
    data1=data.copy()
    data1[feature].hist(bins=10)
    plt.xlabel(feature)
    plt.ylabel("turnover")
    plt.title(feature)
    plt.show()
```
